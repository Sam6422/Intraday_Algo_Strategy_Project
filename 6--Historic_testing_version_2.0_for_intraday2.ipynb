```python
fb = pd.read_excel(r"FO_stocks.xlsx")
# fb = pd.read_excel(r"NSE_Stock_List.xlsx")
b = fb['stock']
buy = []
sell = []
max_loss = 500
for i in range(0,198):
    try:
        data = yf.download(tickers=b[i], period='1d',interval="5m").reset_index()
        data = data.dropna(how='all')
        r = len(data.index)
        low1 = data.iloc[0,3]
        open1 = data.iloc[0,1]
        high1 = data.iloc[0,2]
        order_period = 6
        
        if open1 == low1 and data.iloc[0:order_period,3].min() == low1:
            print(f"open is low {b[i]}")
            if data.iloc[0:order_period,3].min() >= low1:
                units = round(max_loss / (data.iloc[order_period,4] - low1),1)
                max_size = round(units * data.iloc[order_period,4])
                buy.append([b[i],
                       round(data.iloc[order_period,4],2),
                       round(((low1 / data.iloc[order_period,4]) -1) *100 ,2),
                       round(low1,2), units, max_size])
                
        if open1 == high1 and data.iloc[0:order_period,2].max() == high1:
            print(f"open is high {b[i]}")
            if data.iloc[0:order_period,2].max() <= high1:
                units = round(max_loss / (high1 - data.iloc[order_period,4]),1)
                max_size = round(units * data.iloc[order_period,4])
                sell.append([b[i],
                        round(data.iloc[order_period,4],2),
                        round((-(high1 / data.iloc[order_period,4]) +1) *100 ,2),
                        round(high1,2), units, max_size])
        print(f"{i}")
    except:
        pass
        print(f"error in {b[i]}")
```

# 6.) VWAP calculation & implementation

```python
df = yf.download(tickers = 'INDUSINDBK.NS', period = "2d",interval = "5m").reset_index()
df.iloc[75,6] = 147000
# df.iloc[75:85,]
for i in range(0, len(df)):
    df['tp_vol'] = 0
    df['com_tp_vol'] = 0
    df['cum_vol'] = 0
    df['VWAP'] = 0
    
for i in range(75,len(df)):
    tp = (df.iloc[i,2] + df.iloc[i,3] + df.iloc[i,4]) / 3
    tp_v = tp * df.iloc[i,6]
    df.iloc[i,7] = round(tp_v, 2)
    df.iloc[i,8] = df.iloc[75:i,7].sum()
    df.iloc[i,9] = df.iloc[75:i,6].sum()
    df.iloc[i,10] = round(df.iloc[i,8] / df.iloc[i,9],2)
trade=[]
for i in range(75,138):
    if df.iloc[i,4] < df.iloc[i,10] and df.iloc[i-1,4] > df.iloc[i-1,10]:
        trade.append([df.iloc[i,0],'short',
                     df.iloc[i,4],
                     df.iloc[i,10]])
    if df.iloc[i,4] > df.iloc[i,10] and df.iloc[i-1,4] < df.iloc[i-1,10]:
        trade.append([df.iloc[i,0], 'long',
                     round(df.iloc[i,4],2),
                     round(df.iloc[i,10],2)])
```
